```python
import pandas as pd

# Read CSV, skipping rows 0-2, using row 3 as header, and skipping rows 4-5 after that
df = pd.read_csv(
    'be1903.csv',
    encoding='latin1',  # or your working encoding
    skiprows=[0, 1, 2, 4, 5],  # explicitly skip rows by index (0-based)
    header=0  # treat the first non-skipped row (i.e. original row 3) as the header
)

print(df.head())
```

```python
df.columns.values[1] = 'interest_rate_new_business_mortgages'

print(df.columns)
```

```python
import pandas as pd

# Create a monthly datetime index starting in Jan 2003
start_date = '2003-01'
date_range = pd.date_range(start=start_date, periods=len(df), freq='MS')  # 'MS' = Month Start

# Assign this datetime index to the DataFrame
df.index = date_range

df.index.name = 'Date'

print(df.head())
print(df.index)  # Check that it's a DateTimeIndex


```

```python
print(df.columns.tolist())

# Rename using strip to avoid invisible whitespace
df = df.rename(columns={df.columns[1]: 'interest_rate_new_business_mortgages'})

import matplotlib.pyplot as plt

df['interest_rate_new_business_mortgages'].plot(
    figsize=(12, 6),
    title='Interest Rate on New Business Mortgages (Monthly)',
    ylabel='Interest Rate (%)'
)

plt.grid(True)
plt.tight_layout()
plt.show()
```
